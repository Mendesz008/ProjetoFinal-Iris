## Obejtivo

Realizar um análise exploratória do conjnunto de dados Iris para identificar características das diferentes espécies de flores.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data,columns = iris.feature_names)

df['species'] = iris.target

df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Perguntas

## Pergunta 1
Quantas flores existem no conjunto de dados?

```python
df.shape[0]
```

## Pergunta 2
Quantas flores existem de cada espécies?

```python
df['species'].value_counts()
```

```python
df['species'].value_counts().plot(kind='bar')
plt.title('Quantidade de flores por espécie')
plt.show()
```

## Pergunta 3
Qual é a média do comprimento das sépalas?

```python
df['sepal length (cm)'].mean()
```

## Pergunta 4
Qual é a medida da largura das sépalas?

```python
df['sepal width (cm)'].mean()
```

## Pergunta 5
Qual é a média do comprimento das pétalas?


```python
df['petal length (cm)'].mean()
```

## Pergunta 6
Qual é a média da largura das pétalas?

```python
df['petal width (cm)'].mean()
```

## Pergunta 7
Qual a espécie possui o maior comprimento médio de pétala?

```python
df.groupby('species')['petal length (cm)'].mean()
```

```python
df.groupby('species')['petal length (cm)'].mean().plot(kind = 'bar')
plt.title('Comprimento médio das pétalas')
plt.show()
```

## Pergunta 8
Qual espécie possui maior largura média de pétala?

```python
df.groupby('species')['petal width (cm)'].mean()
```

```python
df.groupby('species')['petal width (cm)'].mean().plot(kind ='bar')
plt.title('Largura média das pétalas')
plt.show()
```

A análise exploratória permitiu compreender as características do conjunto de dados Iris. Foi possivel identificar diferenças entre as espécies por meio das medidas das sépalas e pétalas, demonstrando a importância da exploração dos dados antes das análises mais avançadas.
